```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
import os

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data_df = pd.read_csv(clinical_trial_data_to_load)

mouse_drug_data_df.head()
clinical_trial_data_df.head()

```

## Check Mouse Data for Null Values

```python
mouse_drug_data_df.isnull().values.any()

#No null values were returned.
```

## Check Clinical Trial Data for Null Values

```python
clinical_trial_data_df.isnull().values.any()

#No null values were returned.
```

## Merge Data Sources

```python
# Combine the data into a single dataset
merged_df = pd.merge(clinical_trial_data_df, mouse_drug_data_df, on="Mouse ID", how="left")

# Display the data table for preview
merged_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
avg_tumor_volume = merged_df.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()

# Convert to DataFrame, reset index, and rename calculated column to be more representative of info contained
avg_tumor_volume_df = pd.DataFrame(avg_tumor_volume).reset_index()
avg_tumor_volume_df.rename(columns={"Tumor Volume (mm3)": "Avg Tumor Volume (mm3)"}, inplace=True)
                                   
# Preview DataFrame
avg_tumor_volume_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_vol_stderr = merged_df.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem()

# Convert to DataFrame, reset index, and rename calculated column to be more representative of info contained
tumor_vol_stderr_df = pd.DataFrame(tumor_vol_stderr).reset_index()
tumor_vol_stderr_df.rename(columns={"Tumor Volume (mm3)": "Std Error Tumor Volume (mm3)"}, inplace=True)

# Preview DataFrame
tumor_vol_stderr_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_avg_tumor_volume_df = avg_tumor_volume_df.pivot(index="Timepoint",columns='Drug',values="Avg Tumor Volume (mm3)")

# Preview that Reformatting worked
pivot_avg_tumor_volume_df.head()
```

```python
# Generate the Plot (with Error Bars)
plt.figure(figsize=(10,7))

#Create list of drugs for plot

drug_list = [("Capomulin","o","red"),("Infubinol","s","blue"),("Ketapril","^","green"),("Placebo","d","black")]


for drug, markers, colors in drug_list: 
    plt.errorbar(pivot_avg_tumor_volume_df.index, #x-values
                 pivot_avg_tumor_volume_df[drug], #y-values
                 tumor_vol_stderr[drug], #std error
                 marker =  markers, 
                 color = colors,
                 linestyle = '--', 
                 linewidth = 1, 
                 markersize = 7,
                 capthick = 1,
                 capsize = 3)

#Plot customization
plt.legend(("Capomulin", "Infubinol", "Ketapril", "Placebo"), fontsize = 12,loc = "best")
plt.title("Tumor Volume Changes Over Time", fontdict = {"fontsize": 22})
plt.xlabel("Time (Days)", fontdict = {"fontsize": 15})
plt.ylabel("Avg Tumor Volume (mm3)", fontdict = {"fontsize": 15})
plt.grid()

# Save the Figure
plt.savefig(os.path.join("figures","tumor_volume_time.png"))
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
avg_metastatic_site = merged_df.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()

# Convert to DataFrame, reset index, and rename calculated column to be more representative of info contained
avg_metastatic_site_df = pd.DataFrame(avg_metastatic_site).reset_index()
avg_metastatic_site_df.rename(columns={"Metastatic Sites": "Avg # of Metastatic Sites"}, inplace=True)
                                   
# Preview DataFrame
avg_metastatic_site_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_sites_stderr = merged_df.groupby(["Drug", "Timepoint"])["Metastatic Sites"].sem()

# Convert to DataFrame, reset index, and rename calculated column to be more representative of info contained
tumor_vol_stderr_df = pd.DataFrame(metastatic_sites_stderr).reset_index()
tumor_vol_stderr_df.rename(columns={"Metastatic Sites": "Std Error of Metastatic Sites"}, inplace=True)

# Preview DataFrame
tumor_vol_stderr_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_avg_metastatic_site_df = avg_metastatic_site_df.pivot(index="Timepoint",columns='Drug',values="Avg # of Metastatic Sites")

# Preview that Reformatting worked
pivot_avg_metastatic_site_df.head()
```

```python
# Generate the Plot (with Error Bars)
plt.figure(figsize=(10,7))

#Create list of drugs for plot

drug_list = [("Capomulin","o","red"),("Infubinol","s","blue"),("Ketapril","^","green"),("Placebo","d","black")]


for drug, markers, colors in drug_list: 
    plt.errorbar(pivot_avg_metastatic_site_df.index, #x-values
                 pivot_avg_metastatic_site_df[drug], #y-values
                 metastatic_sites_stderr[drug], #std error
                 marker =  markers, 
                 color = colors,
                 linestyle = '--', 
                 linewidth = 1, 
                 markersize = 7,
                 capthick = 1,
                 capsize = 3)

#Plot customization
plt.legend(("Capomulin", "Infubinol", "Ketapril", "Placebo"), fontsize = 12,loc = "best")
plt.title("Metastatic Spread Over Time", fontdict = {"fontsize": 22})
plt.xlabel("Time (Days)", fontdict = {"fontsize": 15})
plt.ylabel("Avg # of Metastatic Sites", fontdict = {"fontsize": 15})
plt.grid()

# Save the Figure
plt.savefig(os.path.join("figures","metastatic_spread_time.png"))
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = merged_df.groupby(["Drug", "Timepoint"])["Mouse ID"].count()

# Convert to DataFrame, reset index, and rename calculated column to be more representative of info contained
mice_count_df = pd.DataFrame(mice_count).reset_index()
mice_count_df.rename(columns={"Mouse ID": "Mouse Count"}, inplace=True)
                                   
# Preview DataFrame
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_mice_count_df = mice_count_df.pivot(index="Timepoint",columns='Drug',values="Mouse Count")

# Preview that Reformatting worked
pivot_mice_count_df.head()
```

```python
# Generate the Plot 
plt.figure(figsize=(10,7))

#Create list of drugs for plot
drug_list = [("Capomulin","o","red"),("Infubinol","s","blue"),("Ketapril","^","green"),("Placebo","d","black")]


for drug, markers, colors in drug_list: 
    plt.plot(pivot_mice_count_df.index, #x-values
                 (pivot_mice_count_df[drug]/pivot_mice_count_df[drug][0])*100, #y-values
                 marker =  markers, 
                 color = colors,
                 linestyle = '--', 
                 linewidth = 1, 
                 markersize = 7)

#Plot customization
plt.legend(("Capomulin", "Infubinol", "Ketapril", "Placebo"), fontsize = 12,loc = "lower left")
plt.title("Survival Rate Over Time", fontdict = {"fontsize": 22})
plt.xlabel("Time (Days)", fontdict = {"fontsize": 15})
plt.ylabel("Survival Rate (%)", fontdict = {"fontsize": 15})
plt.grid()

# Save the Figure
plt.savefig(os.path.join("figures","survival_rate_time.png"))
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
pct_tumor_change = ((pivot_avg_tumor_volume_df.iloc[-1]-pivot_avg_tumor_volume_df.iloc[0])/pivot_avg_tumor_volume_df.iloc[0])*100

# Display the data to confirm
pct_tumor_change
```

```python
# Generate the Plot 
plt.figure(figsize=(10,7))

#Create lists for x and y axis
drug_list = ["Capomulin","Infubinol","Ketapril","Placebo"]
pct_tumor_change_list = [(pct_tumor_change[d])for d in drug_list]

#Color bars by pass vs fail
colors = ["red" if ptc > 0 else 'g' for ptc in pct_tumor_change_list]

#Generate Bar Chart
tumor_change_plt = plt.bar(drug_list, pct_tumor_change_list, color = colors, align = "center")

plt.grid()
plt.axhline(y=0, color = 'black') #adds a horizontal line at zero
plt.ylim(-30,70)
plt.ylabel("Tumor Volume Change(%)", fontsize = 15)
plt.title("Tumor Change over 45 Day Treatment", fontsize = 22)
plt.tick_label = drug_list


# Use functions to label the percentages of changes
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        if height > 0:
            label_position = 2
        else:
            label_position = -8
        plt.text(rect.get_x() + rect.get_width()/2., label_position,
                '%d' % int(height)+'%',color='white',
                ha='center', va='bottom')
                
# Call functions to implement the function calls
autolabel(tumor_change_plt)
        
# Save the Figure
plt.savefig(os.path.join("figures","tumor_change_pct.png"))
```
